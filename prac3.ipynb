2\.Define in your own words the terms constraint, backtracking search, arc consistency,backjumping, min-conflicts, and cycle cutset.

See textbook Chapter6 6.16

_your answer here..._  
constraint: A constraint is a restriction on the possible value of two or more variables.  
backtracking search: backtracking search is a kind of depth-first search. for a certain search tree, the main difference between backtracking and dfs id that backtracking is not in the solution process. the complete tree structure is preserved, while the dfs records the complete search tree.  
they are consistancy. a directed arc from variable A to variable B in a CSP is arc consistent if for every value in the current domain of A, there is some consistent value of B.  
backjumping: backjumping is a way of making backtracking search more efficient. by jumping back more than one level when a dead end is reached.  
min-conflicts is a heuristic for use with local search on CSP problems. the heuristic says that, when given a variable to modify, choose the value that conflicts with the fewest number of the other variables.

```python
from tkinter import *
import time
import random
game = Tk()
game.title('BallGame')
game.geometry("600x600")
game.resizable(False,False)
pad  = Canvas(game,width = 600, height = 600,bd = 0,highlightthickness = 0)
pad.pack()


class Ball:
    x, y = 290, 290
    def __init__(self,pad):         #初始化小球，颜色为红色，直径为20，将其移动到中间
        self.pad = pad
        self.ini = pad.create_oval(0,0,20,20,fill = "red")
        self.pad.move(self.ini,self.x,self.y)



    def move(self):                    #定义小球移动的方法，包括反弹
        global a
        global b
        global paddle
        px = a
        py = b

        if self.x <= 0 or self.x >= 580 and self.y > 0:    #边界反弹
            px = -px + 0.01
            a = px
        elif self.y <= 0:                                   #边界反弹
            py = -py + 0.01
            b = py

        if self.y+20 >= paddle.y and self.y<= paddle.y - py and paddle.x <= self.x +10 <= paddle.x+180 :
            py = -py                                        #滑板反弹
            b = py

        self.pad.move(self.ini, px, py)                     #移动
        self.y = self.y + py
        self.x = self.x + px

    def direction(self):                                 #设置小球最开始移动的方向，随机方向，返回每一步的运动步骤
        horizontal = random.randint(-100 , 100)
        vertical = random.randint(-100 , 100)
        while horizontal ==0 or vertical == 0:
            horizontal = random.randint(-100, 100)
            vertical = random.randint(-100, 100)
        l = (horizontal**2+vertical**2)**0.5
        pathx = 4*horizontal/l
        pathy = 4*vertical/l
        return [pathx,pathy]

    def failure(self):                              #判断游戏是否失败
        if self.y > 580 :
            return True
        else:
            return False


class Paddle:
    x,y = 210, 440
    def __init__(self,pad):      #初始化滑板，颜色为蓝色，移动到画面下方
        self.pad = pad
        self.ini = pad.create_rectangle(0, 0, 180, 20, fill="blue")
        self.pad.move(self.ini, self.x , self.y)

    def moving(self,event):
        x = self.x
        y = self.y
        if 0 <= x <= 420 and 0 <= y <= 580:
            if x == 0:
                if y == 0:
                    if event.keysym == "Down":
                        pad.move(self.ini, 0, 10)
                        y += 10
                        self.y = y
                    if event.keysym == "Right":
                        pad.move(self.ini, 10, 0)
                        x += 10
                        self.x = x
                elif y == 580:
                    if event.keysym == "Up":
                        pad.move(self.ini, 0, -10)
                        y -= 10
                        self.y = y
                    if event.keysym == "Right":
                        pad.move(self.ini, 10, 0)
                        x += 10
                        self.x = x
                else:
                    if event.keysym == "Up":
                        pad.move(self.ini, 0, -10)
                        y -= 10
                        self.y = y
                    elif event.keysym == "Down":
                        pad.move(self.ini, 0, 10)
                        y += 10
                        self.y = y
                    elif event.keysym == "Right":
                        pad.move(self.ini, 10, 0)
                        x += 10
                        self.x = x
            elif x == 420:
                if y == 0:
                    if event.keysym == "Down":
                        pad.move(self.ini, 0, 10)
                        y += 10
                        self.y = y
                    if event.keysym == "Left":
                        pad.move(self.ini, -10, 0)
                        x -= 10
                        self.x = x
                elif y == 580:
                    if event.keysym == "Up":
                        pad.move(self.ini, 0, -10)
                        y -= 10
                        self.y = y
                    if event.keysym == "Left":
                        pad.move(self.ini, -10, 0)
                        x -= 10
                        self.x = x
                else:
                    if event.keysym == "Up":
                        pad.move(self.ini, 0, -10)
                        y -= 10
                        self.y = y
                    elif event.keysym == "Down":
                        pad.move(self.ini, 0, 10)
                        y += 10
                        self.y = y
                    elif event.keysym == "Left":
                        pad.move(self.ini, -10, 0)
                        x -= 10
                        self.x = x
            else:
                if y == 0:
                    if event.keysym == "Right":
                        pad.move(self.ini, 10, 0)
                        x += 10
                        self.x = x
                    if event.keysym == "Down":
                        pad.move(self.ini, 0, 10)
                        y += 10
                        self.y = y
                    if event.keysym == "Left":
                        pad.move(self.ini, -10, 0)
                        x -= 10
                        self.x = x
                elif y == 580:
                    if event.keysym == "Right":
                        pad.move(self.ini, 10, 0)
                        x += 10
                        self.x = x
                    if event.keysym == "Up":
                        pad.move(self.ini, 0, -10)
                        y -= 10
                        self.y = y
                    if event.keysym == "Left":
                        pad.move(self.ini, -10, 0)
                        x -= 10
                        self.x = x
                else:
                    if event.keysym == "Right":
                        pad.move(self.ini, 10, 0)
                        x += 10
                        self.x = x
                    if event.keysym == "Up":
                        pad.move(self.ini, 0, -10)
                        y -= 10
                        self.y = y
                    elif event.keysym == "Down":
                        pad.move(self.ini, 0, 10)
                        y += 10
                        self.y = y
                    elif event.keysym == "Left":
                        pad.move(self.ini, -10, 0)
                        x -= 10
                        self.x = x
        else:
            pass             #滑板移动方法   


ball = Ball(pad)
paddle = Paddle(pad)
di = ball.direction()
a = di[0]
b = di[1]
pad.bind_all("<KeyPress-Up>", paddle.moving)
pad.bind_all("<KeyPress-Down>", paddle.moving)
pad.bind_all("<KeyPress-Right>", paddle.moving)
pad.bind_all("<KeyPress-Left>", paddle.moving)
while True:
    ball.move()
    if ball.failure():
        break
    game.update_idletasks()
    game.update()
    time.sleep(0.00001)

game.mainloop()


```
